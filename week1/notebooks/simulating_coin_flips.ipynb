```python
n = 100_000
heads_so_far = 0
fraction_of_heads = []
for i in range(n):
    if cpi.flip_fair_coin() == 'heads':
        heads_so_far += 1
    fraction_of_heads.append(heads_so_far / (i+1))

plt.figure(figsize=(8,4))
plt.plot(range(1, n+1), fraction_of_heads);

plt.xlabel('Number of flips')
plt.ylabel('Fraction of heads');
```
